# Tokenizers

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("sshleifer/distilbart-cnn-12-6")
```

```python
raw_inputs =     [
        "I love deep learning!",
        "I hate this so much!",
    ]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
print(inputs)
```

```python
print('Tokenizer output for "I love deep learning!"')
print(f"Input ids: {inputs['input_ids'][0]}")
print(f"Attention Mask: {inputs['attention_mask'][0]}")
print("-"*100)
print('Tokenizer output for "I hate this so much!"')
print(f"Input ids: {inputs['input_ids'][1]}")
print(f"Attention Mask: {inputs['attention_mask'][1]}")
```

```python
tokenizer('I love deep learning!')
```

### Tokenizers Under The Hood

```python
tokens = tokenizer.tokenize('I love deep learning!')
tokens
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
token_ids
```

```python
decoded_tokens = tokenizer.decode(token_ids)
decoded_tokens
```

```python
model_prepped_ids = tokenizer.prepare_for_model(token_ids)
model_prepped_ids
```

# Pipeline

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(
    [
        "I love deep learning!",
        "I hate this so much!",
    ]
)
```

```python
classifier(
    [
        "Generative AI is fun!",
        "Mathematics can be boring sometimes.!",
    ]
)
```

```python
text_generator = pipeline("text-generation")

text_generator([
    'I love Mathematics but',
    'All the monarchies in the world were affected by Industrial Revolution'
])
```

```python
summarizer = pipeline("summarization")

summarizer([
    """

    HyDE (Hierarchical Document Embedding) helps in capturing both global and local context in documents, improving search relevance. MMR (Maximal Marginal Relevance) balances relevance and diversity in search results, which I used to enhance the quality of retrieved documents. LLM reranking involves using a fine-tuned LLM to re-rank search results based on their semantic relevance. In a semantic search project, these techniques improved the accuracy and diversity of search results, making the system more useful for users.

    """
])
```

## Accessing Pretrained models

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
inputs = tokenizer('I like Generative AI', return_tensors='pt')
inputs
```

```python
outputs = model(**inputs)
outputs
```

## Model Embeddings

```python
from transformers import AutoModel

model = AutoModel.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
inputs = tokenizer('I like Generative AI', padding=True, truncation=True, return_tensors='pt')

outputs = model(**inputs)

print(outputs.last_hidden_state.shape) # the token embeddings
```

```python
# to get the full context vector for the sequence
context_vectors = outputs.last_hidden_state.mean(dim=1)
context_vectors.shape
```

## Accessing Model Config & Creating Custom Models
